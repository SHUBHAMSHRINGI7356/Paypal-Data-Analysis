```python

transactions['transactions_date'] = pd.to_datetime(transactions['transaction_date'])
users['signup_date'] = pd.to_datetime(users['signup_date'])
chargebacks['chargeback_date'] = pd.to_datetime(chargebacks['chargeback_date'])
                                                

```

```python
#Enables:
#Monthly analysis
#Time difference calculations
#Date logic is critical in payments analytics
```

```python
# Remove Invalid Transactions
```

```python
#Remove Duplicate Transactions
```

```python
transactions = transactions.drop_duplicates(subset = 'transaction_id')
```

```python
#Standardize Categorical Values
```

```python
transactions['transaction_status'] = (transactions['transaction_status'].str.strip().str.title())
```

```python
#Now we create business features.
```

```python
#Feature 1: Is Successful Transaction
transactions['IS_Success']= transactions['transaction_status'] == 'Success'
```

```python
print(transactions['transaction_date'].dtype)
print(transactions['transaction_date'].head())
```

```python
transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'])
```

```python
print(transactions['transaction_date'].head())
```

```python
#Transaction Month
transactions['txn_month'] = transactions['transaction_date'].dt.to_period('M')
transactions['txn_month_name'] = transactions['transaction_date'].dt.month_name()
```

```python
print(transactions['txn_month_name'].head())
```

```python
#Join Payment Method Fees
```

```python
transactions = transactions.merge( payment_methods, on ='payment_method', how = 'left')
```

```python
transactions.head()
```

```python
transactions = transactions.drop(columns=['IS_Successfull'])
```

```python
transactions.head(10)
```

```python
#Processing Fee Amount
transactions['proccessing_fee'] = (transactions['amount'] * transactions['processing_fee_pct'] / 100)
```

```python
#Net Revenue
transactions['net_revenue'] = transactions['amount'] - transactions['proccessing_fee']
```

```python
transactions['net_revenue'].head()
```

```python
#Chargeback Flag
transactions['is_chargeback'] = transactions['transaction_id'].isin(chargebacks['transaction_id'])
```

```python
transactions.head(10)
```

```python
#User-Level Enrichment
transactions = transactions.merge( users[['user_id', 'country', 'account_type']] , on= 'user_id', how = 'left')
```

```python
#Final Validation
transactions[['amount','proccessing_fee','net_revenue']].describe()
transactions.isnull().sum()
```
